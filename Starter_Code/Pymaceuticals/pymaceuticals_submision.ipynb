### Analysis

-The main purpose of this analysis was to help Pymaceuticals, Inc., a pharmaceutical company that specializes in anti-cancer medications. to compare the effectiveness of the use of Capomulin against other drugs used to treat squamous cell carcinoma (SCC), a commonly occurring form of skin cancer.
In this study, 249 mice who were identified with SCC tumors received treatment with a range of drug regimens. Over the course of 45 days, tumor development was observed and measured.

Observations:

1- Mouses treated with Capomulin had on average smaller tumor volume compared with the other treatments, except for Ramicane that had a slightly lower average tumor volume.

2- The number of timepoints observed with Capomulin were higher than compared with the other drugs.

3- There is positive correlation between the mouse weight and tumor volume, meaning the higher the mouse's weight the bigger the tumor

4- Seems like Capomulin is the best treatment option for this type of tumor, followed very closely by Ramicane. More studies on these two are suggested.


 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_df = pd.merge(study_results,mouse_metadata,how = "left", on="Mouse ID")

# Display the data table for preview
study_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
mean_vol = new_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median_vol = new_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
variance_vol = new_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
std_vol = new_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
SEM_vol = new_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_data= pd.DataFrame({'Mean Tumor Volume': mean_vol,
                              'Median Tumor Volume' : median_vol,
                              'Tumor Volume Variance': variance_vol,
                              'Tumor Volume Std. Dev.' : std_vol,
                              'Tumor Voluma Std. Err.':SEM_vol})
summary_data
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_table = new_study_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ['sum', 'mean', 'var', 'std','sem']})
summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint = new_study_df.groupby("Mouse ID")['Timepoint'].max()
greatest_timepoint = greatest_timepoint.reset_index() 


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = greatest_timepoint.merge(new_study_df, on=['Mouse ID', 'Timepoint'], how='left')
merged_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin','Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
     
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_volume = merged_data.loc[merged_data['Drug Regimen'] == drug, "Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol_data.append(tumor_volume)
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_volume.quantile([.25,.5,.75])
    Q1 = quartiles[.25]
    Q3 = quartiles[.75]
    IQR = Q3 - Q1
    lower_end = Q1 - 1.5 * IQR
    upper_end = Q3 + 1.5 * IQR
    outliers = tumor_volume.loc[(tumor_volume < lower_end) | (tumor_volume > upper_end)]
    print(f'({drug} potential outliers: {outliers})')                                                                     
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
outlier_marker = dict(marker='o', markersize=10, markerfacecolor='red', markeredgecolor='black')
plt.boxplot(tumor_vol_data, labels = treatments, flierprops= outlier_marker)
plt.title('Distribution Of Final Tumor Volume Per Treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.xlabel('Treatments')

plt.show()
```
